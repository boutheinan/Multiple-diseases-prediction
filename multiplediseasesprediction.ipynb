```python
import pandas as pd

train_df = pd.read_csv("Blood_samples_dataset_balanced_2(f).csv")  #
test_df = pd.read_csv("blood_samples_dataset_test.csv")


train_df.head()
```

```python
X_train = train_df.drop("Disease", axis=1)
y_train = train_df["Disease"]

X_test = test_df.drop("Disease", axis=1)
y_test = test_df["Disease"]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

model = RandomForestClassifier(random_state=42)
model.fit(X_train, y_train)

predictions = model.predict(X_test)

print(classification_report(y_test, predictions))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

importances = model.feature_importances_
features = X_train.columns


plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features)
plt.title("Important Blood Parameters in Disease Prediction")
plt.xlabel("Importance Score")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Get the importance scores for each feature
importances = model.feature_importances_

# List of feature names (blood parameters)
features = X_train.columns

# Create a bar plot to show the importance of each blood parameter
plt.figure(figsize=(10, 6))
sns.barplot(x=importances, y=features)
plt.title("Importance of Blood Parameters in Disease Prediction")
plt.xlabel("Importance Score")
plt.ylabel("Blood Parameters")
plt.tight_layout()
plt.show()
```

```python
# Count how many times each disease was predicted
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.countplot(x=predictions)
plt.title("Predicted Disease Distribution")
plt.xlabel("Disease")
plt.ylabel("Count of Predictions")
plt.tight_layout()
plt.show()
```

```python
# Predict on the test set
predictions = model.predict(X_test)
```

```python
# Compare predictions with actual values (y_test)
comparison = pd.DataFrame({'Actual': y_test, 'Predicted': predictions})
print(comparison.head())
```

```python
# Extract metrics from classification report
report = classification_report(y_test, predictions, output_dict=True)

# Create a DataFrame from the classification report
report_df = pd.DataFrame(report).transpose()

# Plot the classification report metrics for each disease
report_df.plot(kind='bar', figsize=(10, 6), colormap='viridis')
plt.title("Classification Report: Precision, Recall, and F1-Score")
plt.xlabel("Disease Classes")
plt.ylabel("Scores")
plt.tight_layout()
plt.show()

```
